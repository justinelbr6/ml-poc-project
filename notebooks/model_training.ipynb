# Entraînement des Modèles

Nous allons tester 3 algorithmes en optimisant la métrique **Recall** via une validation croisée (GridSearch).

```python
import sys
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.model_selection import GridSearchCV

sys.path.append("..")
from src.data import load_dataset_split
from src.config import MODEL_PARAMS

X_train, X_test, y_train, y_test = load_dataset_split()
print(f"Jeu d'entrainement: {X_train.shape}, Jeu de test: {X_test.shape}")
```

## Lancement des modèles

```python
models = {
    "LogisticRegression": LogisticRegression(random_state=42),
    "RandomForest": RandomForestClassifier(random_state=42),
    "GradientBoosting": GradientBoostingClassifier(random_state=42)
}

best_estimators = {}
for name, model in models.items():
    print(f"\n⚙️ Optimisation de {name}...")
    grid = GridSearchCV(estimator=model, param_grid=MODEL_PARAMS[name], cv=5, scoring='recall', n_jobs=-1)
    grid.fit(X_train, y_train)
    best_estimators[name] = grid.best_estimator_
    print(f"✅ Meilleurs paramètres: {grid.best_params_}")
```
