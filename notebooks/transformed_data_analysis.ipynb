# Analyse Post-Transformation

Ce notebook prouve visuellement l'intérêt de nos transformations, notamment la barrière non-linéaire.

```
import sys
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

sys.path.append("..")
from src.data import load_and_preprocess_data

X, y = load_and_preprocess_data("../data/Heart_disease_cleveland_new.csv")
df_full = X.copy()
df_full['target'] = y
```

## 1. Impact sur la corrélation

```
plt.figure(figsize=(10, 8))
target_corr = df_full.corr()['target'].drop('target').sort_values()
target_corr.plot(kind='barh', color='teal')
plt.title('Corrélation des features transformées avec la maladie')
plt.show()
```

## 2. Limite Linéaire via l'Analyse en Composantes Principales (PCA)
Ce graphe montre pourquoi on ne peut pas utiliser une simple ligne pour séparer les malades des sains.

```
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

plt.figure(figsize=(10, 8))
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=y, alpha=0.7, palette='Set1')
plt.title('Projection PCA (2D) - Enchevêtrement des classes')
plt.show()
```
