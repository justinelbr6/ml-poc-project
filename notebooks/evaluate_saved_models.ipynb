# Évaluation Clinique des Modèles

On évalue la précision et surtout le **Recall** (Rappel) des modèles pour éviter les faux négatifs.

```
import sys
import joblib
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, roc_curve, precision_recall_curve

sys.path.append("..")
from src.data import load_dataset_split

X_train, X_test, y_train, y_test = load_dataset_split()

models = {
    "Régression Logistique": joblib.load("../models/logistic_regression.pkl"),
    "Random Forest": joblib.load("../models/random_forest.pkl"),
    "Gradient Boosting": joblib.load("../models/gradient_boosting.pkl")
}
print("🚀 Modèles chargés avec succès.")
```

## 1. Matrices de Confusion
Objectif : Réduire à zéro le cadran haut-droite (Faux Négatifs).

```
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
for ax, (name, model) in zip(axes, models.items()):
    y_pred = model.predict(X_test)
    cm = confusion_matrix(y_test, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm)
    disp.plot(cmap='Blues', ax=ax, colorbar=False)
    ax.set_title(f'Matrice : {name}')
    ax.grid(False)
plt.show()
```

## 2. Explicabilité (Feature Importance)

```
import pandas as pd
import seaborn as sns

gb_model = models.get('Gradient Boosting')
if gb_model and hasattr(gb_model, 'feature_importances_'):
    importances = gb_model.feature_importances_
    df_imp = pd.DataFrame({'Feature': X_test.columns, 'Importance': importances})
    df_imp = df_imp.sort_values(by='Importance', ascending=False).head(10)
    
    plt.figure(figsize=(10, 6))
    sns.barplot(data=df_imp, x='Importance', y='Feature', palette='magma')
    plt.title('Top 10 Variables Cliniques (Gradient Boosting)')
    plt.show()
```
